```python
## Task 1: Mark all launch sites on a map

# --- Step 1: Load Data ---
# Note: You must ensure this file is loaded in your environment.
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
spacex_df = pd.read_csv('spacex_launch_geo.csv')


# --- Step 2: Extract Unique Launch Sites ---
# Get only the columns needed and group to find unique sites and their coordinates
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
print(launch_sites_df)

# --- Step 3: Initialize the Map ---
# Define a central location (e.g., Cape Canaveral) for map centering
# For CCAFS SLC 40: [28.561857, -80.577366]
center_coords = [28.561857, -80.577366]

# Create a Folium map object centered on a launch site location
site_map = folium.Map(location=center_coords, zoom_start=5)

# --- Step 4: Add Markers for Each Site ---
for index, site in launch_sites_df.iterrows():
    # A. Add a CircleMarker for the site location
    folium.CircleMarker(
        location=[site['Lat'], site['Long']],
        radius=10,  # size of the circle
        color='#000000', # Border color
        fill=True,
        fill_color='#d35400', # Fill color
        fill_opacity=0.8,
        popup=site['Launch Site']
    ).add_to(site_map)
    
    # B. Add a DivIcon Marker for the site name
    folium.Marker(
        location=[site['Lat'], site['Long']],
        icon=DivIcon(
            icon_size=(200, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12pt; color:#d35400;"><b>{site["Launch Site"]}</b></div>',
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities

import math
from folium.features import DivIcon

# Step 1: Define Haversine distance function
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    return R * c


# Step 2: Select one launch site (example: CCAFS SLC-40)
launch_site = launch_sites_df[launch_sites_df['Launch Site'] == 'CCAFS SLC-40']

launch_lat = launch_site['Lat'].values[0]
launch_lon = launch_site['Long'].values[0]


# Step 3: Define example proximity coordinates
# (Approximate values are acceptable for the lab)

coastline_lat, coastline_lon = 28.5623, -80.5678
highway_lat, highway_lon = 28.5721, -80.5853
city_lat, city_lon = 28.5383, -81.3792     # Orlando
railway_lat, railway_lon = 28.5720, -80.5850


# Step 4: Calculate distances
distance_coastline = calculate_distance(launch_lat, launch_lon, coastline_lat, coastline_lon)
distance_highway = calculate_distance(launch_lat, launch_lon, highway_lat, highway_lon)
distance_city = calculate_distance(launch_lat, launch_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_lat, launch_lon, railway_lat, railway_lon)


# Step 5: Draw distance lines and labels on the map
proximities = [
    ("Coastline", coastline_lat, coastline_lon, distance_coastline),
    ("Highway", highway_lat, highway_lon, distance_highway),
    ("City", city_lat, city_lon, distance_city),
    ("Railway", railway_lat, railway_lon, distance_railway)
]

for name, lat, lon, distance in proximities:
    
    # Draw line
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], [lat, lon]],
        weight=2
    ).add_to(site_map)
    
    # Add distance label
    folium.Marker(
        location=[lat, lon],
        icon=DivIcon(
            icon_size=(200, 36),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px;"><b>{name}:</b> {distance:.2f} km</div>'
        )
    ).add_to(site_map)

# Display final map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
